#### 유사도(Proximity)

1. Euclidean distance

2. Cosine Similarity

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
v1 = np.array([1,2,3])
v2 = np.array([0,0,0])
````

넘파이는 수치 연산을 위한 모듈이기 때문에 당연히 넘파이 연산은 숫자연산 즉, 벡터연산이 잘 되어야함

````python
L = [1,2,3]
L2 = [0,0,0]

# List 각 원소들끼리 값을 더하는 코드 == vector addition과 동일한 결과
for i, i2 in zip(L,L2) :
    print(i+i2)

L + L2 # concatenation
````

````python
# numpy boradcasting

v1 + v2 # vector addition # (1,2,3) + (0,0,0) = (1,2,3)
v2 - v1 # vector subtraction
v1 * v2 # elementwise multiplication # 원소 by 원소로 곱하기 연산을 해준것 즉, boradcasting한 것
v2 / v1 # elementwise division
v1 + 3 # 3을 확장하여 boradcasting 함
v1 * 2 # scalar multiplication


## universal function(numpy에만 존재하는 개념) - numpy에서 넘어온 universal function 적용!
def reverse_num(x):
    return 1/x

# boradcasting
for i in v1:
    print(reverse_num(i)) 
# 속도 차이 1000배정도 남, for문은 최대한 안쓰는게 좋음
# 모듈에 구현되어 있는 코드(만들어진 함수)를 사용하는 것이 좋음, 그것이 맞는 방법임
# 직접 코드를 만들어서 해보고 싶으면 C++을 사용하는 것이 좋음!

1 / v1
````

- 브로드캐스팅을 함수의 개념으로 가져와서 확장한 개념이 universal function!!

- 속도 차이 1000배정도 남, for문은 최대한 안쓰는게 좋음
- 모듈에 구현되어 있는 코드를 만들어진 개념을 사용하는 것이 좋음, 그것이 맞는 방법임
- 직접 코드를 만들어서 해보고 싶으면 C++을 사용하는 것이 맞음!


````python
## Euclidean distance(L2 distance) = v1과 v2에 있는 모든 원소들끼리 뺀 다음에, 제곱하고 다 더해서 루트 씌운거 (최단거리)

# (v1 - v2) * (v1 - v2)
# (v1 - v2) **2

distance = np.sqrt(np.sum(np.square(v1-v2)))

distance2 = np.linalg.norm(v1 - v2) # 벡터의 크기를 구하는! default - L2 norm

distance == distance2
````

````python
# lambda를 이용하여 코드 한줄로 L2 distance를 표현
f = lambda x,y : np.sqrt(np.sum(np.square(x - y)))
f(v1, v2)
````

````python
# universal function 하여 브로드캐스팅을 함수의 개념으로 확장한 것
def euclidean_distance(x,y):
    return np.sqrt(np.sum(np.square(x - y)))

euclidean_distance(v1, v2)


def euclidean_distance2(x,y):
    return np.linalg.norm(x-y)

euclidean_distance2(v1, v2)
````

````python
def cosine_similarity(a, b):
    # 1. 분자 : a o b
    dot = a @ b
    
    # 2. 분모 : ||a||, ||b||
    norm = np.linalg.norm(a) * np.linalg.norm(b) + 1e-6 # 1e-12 이상은 비추천
    return dot / norm
````

내적 연산자
- numpy : dot 연산 지원 (  np.dot(a,b)  )
- python 3.5 이상부터 : @ 연산자 지원 (a @ b)

````python
cosine_similarity(v1,v2) # 0 division이라서 당연히 안됨

# 코사인 시밀러리티 절대값이 높은게 유사성이 높다.
````
