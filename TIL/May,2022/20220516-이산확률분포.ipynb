```python
x_set = np.array([1,2,3,4,5,6])
```

```python
prob = np.array([f(x_k) for x_k in x_set])
```

```python
# 확률질량함수 정의
def f(x) :
    if x in x_set :
        return x/21
    else :
        return 0
    
# 누적분포함수 정의
def F(x) :
    return np.sum([f(x_k) for x_k in x_set if x_k <= x])
```

```python
# 통계 계산 함수 1
# E(X) 확률변수의 기댓값을 계산하는 함수

def E(X, g=lambda x:x) :
    x_set, f = X
    return np.sum([g(x_k) * f(x_k) for x_k in x_set])
```

```python
# 통계 계산 함수 2
# V(X) 확률변수의 분산을 계산하는 함수

def V(X, g=lambda x:x) :
    x_set, f = x
    mean = E(X, g)
    return np.sum([(g(x_k) - mean)** 2 * f(x_k) for x_k in x_set])
```

```python
# 통계 계산 함수 3
# check_prob 확률변수의 성질을 만족하는지 검정하는 함수
# 1. 확률이 절대적으로 0 이상인지 확인 2. 모든 확률 더해서 1인지 확인 3. 기대값, 분산 확인

def check_prob(X) :
    x_set, f = x
    prob = np.array([f(x_k) for x_k in x_set])
    assert np.all(prob >= 0), 'minus probability'
    prob_sum = np.round(np.sum(prob), 6)
    assert prob_sum == 1, f'sum of probability{prob_sum}'
    print(f'expected value {E(X) : .4}')
    print(f'variance {(V(X)):.4}')
    
# assert() 가정설정문 : 방어적 프로그래밍에서 사용하여 코드를 점검하는데 사용, 어떤 조건을 테스트하는 디버깅 보조 도구
# 뒤의 조건이 True가 아니면 AssertError 발생
```

```python
# 통계 계산 함수 4
# plot_prob 확률변수의 확률밀도함수 및 기대값을 시각화

def plot_prob(X) :
    x_set, f = X
    prob = np.array([f(x_k) for x_k in x_set])
    
    fig = plt.figure(figsize=(10,6))
    ax = fig.add_subplot(1,1,1)
    ax.bar(x_set, prob, label = 'prob')
    ax.vlines(E(X), 0, 1, label = 'mean')
    ax.set_xticks(np.append(x_set,E(X)))
    ax.set_ylim(0, prob.max()*1, 2)
    ax.legend()
    
    plt.show()
```

### 베르누이 분포(Bernoulli Distribution)

```python
# numpy 활용하여 베르누이 분포 확인

def Bern(p) :
    x_set = np.array([0,1])
    def f(x) :
        if x in x_set :
            return p ** x * (1-p) ** (1-x)
        else :
            return 0
    return x_set, f
```

```python
# X~Bern(0.3) 호출 및 기대값/분산 계산해보기 
p = 0.3
x = Bern(p)
check_prob(x)
```

```python
plot_prob(x)
```

```python
# Scipy.stats의 Bernoulli 함수 활용

rv = stats.bernoulli(p)
```

```python
# pmf : 확률질량함수
rv.pmf(0), rv.pmf(1)
```

```python
rv.pmf([0,1])
```

```python
# cdf : 누적분포함수
rv.cdf([0,1])
```

```python
rv.mean(), rv.var()
```

### 이항분포(Binomial Distribution)

```python
# numpy를 활용하여 구현

from scipy.special import comb

def Bin(n, p) :
    x_set = np.arange(n+1)
    def f(x):
        if x in x_set :
            return comb(n, x) * p**x * (1-p)**(n-x)
        else : 
            return 0
    return x_set,f
```

```python
# X~Bin(10, 0.3) 
n = 10
p = 0.3
x = Bin(n, p)
```

```python
check_prob(x)
```

```python
plot_prob(x)
```

```python
# scipy.stats의 binom함수 활용

fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot(111)

x_set = np.arange(n+1)
for p in [0.3, 0.5, 0.7] :
    rv = stats.binom(n, p)
    ax.plot(x_set, rv.pmf(x_set), label = f'p:{p}', color ='gray')
ax.set_xticks(x_set)
ax.legend()

plt.show()
```

### 기하분포 (Geometric Distribution)

```python
# numpy를 활용하여 X~Ge(p)구현
def Ge(p) :
    x_set = np.arange(1, 30)
    def f(x) :
        if x in x_set :
            return (1-p)**(x-1) * p
        else : 
            return 0
    return x_set, f
```

```python
# X~Ge(0.5) 기댓값/분산 계산 및 시각화

p = 0.5
X = Ge(p)
check_prob(X)
```

```python
plot_prob(X)
```

```python
# Scipy.stats의 geom함수 사용

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

x_set = np.arange(1,15)
for p in [0.2, 0.5, 0.8] :
    rv = stats.geom(p)
    ax.plot(x_set, rv.pmf(x_set), label = f'p:{p}', color = 'gray')
    
ax.set_xticks(x_set)
ax.legend()

plt.show()
```

### 푸아송 분포(Poisson Distribution)

```python
# numpy 활용하여 X~Poi(λ) 표현

from scipy.special import factorial

def Poi(lam) :
    x_set = np.arange(20)
    def f(x) :
        if x in x_set :
            return np.power(lam,x) / factorial(x) * np.exp(-lam)
        else : 
            return 0
    return x_set, f
```

```python
# X~Poi(3) 기댓값/분산 계산 및 시각화

lam = 3
X = Poi(lam)
```

```python
check_prob(X)
```

```python
plot_prob(X)
```

```python
# Scipy.stats의 poission함수 활용

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

x_set = np.arange(20)
for lam in [3,5,8] :
    rv = stats.poisson(lam)
    ax.plot(x_set, rv.pmf(x_set), label=f'lam:{lam}', color='gray')
    
ax.set_xticks(x_set)
ax.legend()

plt.show()
```
